```python
import math
from collections import Counter, defaultdict
import numpy as np
import matplotlib.pyplot as plt
import dendropy

def canon_label(x: str) -> str:
    return x.strip().strip("'\"").replace(" ", "_")

def relabel_tree_inplace(tree):
    for nd in tree.leaf_node_iter():
        if nd.taxon is not None and nd.taxon.label is not None:
            nd.taxon.label = canon_label(nd.taxon.label)

def topology_key_unrooted(tree):
    """
    Returns a canonical tuple key for the unrooted topology using non-trivial splits.
    Assumes tree.taxon_namespace is consistent across trees.
    """
    tree.is_rooted = False
    tree.encode_bipartitions()
    full_mask = tree.taxon_namespace.all_taxa_bitmask()
    splits = []
    for b in tree.bipartition_encoding:
        if b.is_trivial():
            continue
        x = b.split_bitmask
        x = min(x, full_mask ^ x)  # canonical orientation
        splits.append(x)
    return tuple(sorted(splits))

def load_treelist(path, schema, taxa):
    trees = dendropy.TreeList.get(
        path=path,
        schema=schema,
        taxon_namespace=taxa,
        preserve_underscores=True,
    )
    for t in trees:
        t.is_rooted = False
        relabel_tree_inplace(t)
    return trees

def topology_mass_from_trees(trees, weighted=False):
    """
    Returns dict: topo_key -> probability mass.
    If weighted=True, uses t.weight when present (MrBayes trprobs).
    """
    counts = defaultdict(float)
    total = 0.0
    for t in trees:
        w = float(t.weight) if (weighted and getattr(t, "weight", None) is not None) else 1.0
        counts[topology_key_unrooted(t)] += w
        total += w
    return {k: v/total for k, v in counts.items()} if total > 0 else {}

# paths
sampled_nexus_path = "results_converge_elbo/run_20260219_001447_ntaxa27/sampled_trees.nex"
trprobs_path = "data/DS1.trprobs"

taxa = dendropy.TaxonNamespace()
sampled = load_treelist(sampled_nexus_path, "nexus", taxa)
mb = load_treelist(trprobs_path, "nexus", taxa)

Q = topology_mass_from_trees(sampled, weighted=False)  # your model
P = topology_mass_from_trees(mb, weighted=True)        # MrBayes

print("Unique topologies (model):", len(Q))
print("Unique topologies (MB):", len(P))

overlap = set(P) & set(Q)
print("Support overlap count:", len(overlap))
print("MB mass on overlap:", sum(P[k] for k in overlap))
print("Model mass on overlap:", sum(Q[k] for k in overlap))
```

```python
# Plot top‑K topology masses (side-by-side)
def topk_items(d, k=20):
    return sorted(d.items(), key=lambda kv: kv[1], reverse=True)[:k]

K = 1000
topP = topk_items(P, K)
topQ = topk_items(Q, K)

# Use ranks 1..K; labels are just rank indices (topology keys are huge)
x = np.arange(1, K+1)

pvals = [v for _, v in topP] + [0.0]*max(0, K-len(topP))
qvals = [v for _, v in topQ] + [0.0]*max(0, K-len(topQ))

plt.figure(figsize=(10,4))
plt.bar(x - 0.2, pvals[:K], width=0.4, label="MrBayes (top-K)")
plt.bar(x + 0.2, qvals[:K], width=0.4, label="Model (top-K)")
plt.xlabel("Topology rank (within each distribution)")
plt.ylabel("Probability mass")
plt.title(f"Top-{K} Topology Masses (Unrooted)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
def rank_curve(d):
    vals = sorted(d.values(), reverse=True)
    ranks = np.arange(1, len(vals)+1)
    return ranks, np.array(vals)

rP, vP = rank_curve(P)
rQ, vQ = rank_curve(Q)

plt.figure(figsize=(6,4))
plt.loglog(rP, vP, label="MrBayes")
plt.loglog(rQ, vQ, label="Model")
plt.xlabel("Rank")
plt.ylabel("Mass")
plt.title("Topology Rank–Mass (log-log)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import math
from collections import defaultdict, Counter
import dendropy

# ---------- paths ----------
trprobs_path = "data/DS1_synth_1000/seq_gen_DS1_for_mb.nex.trprobs"
sampled_path = "results_ca_vi/run_20260226_170508_ntaxa27/sampled_trees.nwk"  # or .nex

# ---------- helpers ----------
def canon_label(x: str) -> str:
    return x.strip().strip("'\"").replace(" ", "_")

def relabel_tree_inplace(tree):
    for nd in tree.leaf_node_iter():
        if nd.taxon is not None and nd.taxon.label is not None:
            nd.taxon.label = canon_label(nd.taxon.label)

def topology_key_unrooted(tree):
    tree.is_rooted = False
    tree.encode_bipartitions()
    full_mask = tree.taxon_namespace.all_taxa_bitmask()
    splits = []
    for b in tree.bipartition_encoding:
        if b.is_trivial():
            continue
        x = b.split_bitmask
        x = min(x, full_mask ^ x)   # canonical orientation
        splits.append(x)
    return tuple(sorted(splits))

def normalize(d):
    s = sum(d.values())
    return {k: v/s for k, v in d.items()} if s > 0 else {}

# ---------- load trees with shared namespace ----------
taxa = dendropy.TaxonNamespace()

mb = dendropy.TreeList.get(
    path=trprobs_path, schema="nexus", taxon_namespace=taxa, preserve_underscores=True
)

# if your sampled file is Newick (one tree per line):
model = dendropy.TreeList.get(
    path=sampled_path, schema="newick", taxon_namespace=taxa, preserve_underscores=True
)
# if it's nexus, use schema="nexus"

for t in mb:
    relabel_tree_inplace(t)
for t in model:
    relabel_tree_inplace(t)

# ---------- P_MB from weights ----------
p_counts = defaultdict(float)
for t in mb:
    w = float(t.weight) if getattr(t, "weight", None) is not None else 1.0
    p_counts[topology_key_unrooted(t)] += w
P = normalize(p_counts)

# ---------- Q_model from frequencies ----------
q_counts = Counter(topology_key_unrooted(t) for t in model)
Nq = sum(q_counts.values())

# ---------- KL(P||Q) with smoothing on MB support ----------
eps = 1e-10
den = Nq + eps * len(P)

kl = 0.0
for topo, p in P.items():
    q = (q_counts.get(topo, 0) + eps) / den
    kl += p * (math.log(p) - math.log(q))

# diagnostics
overlap = sum(1 for k in P if k in q_counts)
overlap_mass = sum(P[k] for k in P if k in q_counts)

print("MB unique topologies:", len(P))
print("Model unique topologies:", len(q_counts))
print("Support overlap:", overlap)
print("P_MB mass on overlap:", overlap_mass)
print("KL(P_MB || Q_model):", kl)
```

```python
import math
from collections import Counter
import numpy as np
import dendropy
from dendropy.calculate import treecompare

# ---------- paths ----------

trprobs_path = "data/DS1/DS1_for_mb.nex.trprobs"
sampled_path = "results_ca_vi/run_20260305_142949_ntaxa27/sampled_trees.nwk"  # change if needed

# ---------- helpers ----------
def canon_label(x: str) -> str:
    return x.strip().strip("'\"").replace(" ", "_")

def relabel_tree_inplace(tree):
    for nd in tree.leaf_node_iter():
        if nd.taxon is not None and nd.taxon.label is not None:
            nd.taxon.label = canon_label(nd.taxon.label)

def leaf_set(tree):
    return {nd.taxon.label for nd in tree.leaf_node_iter() if nd.taxon is not None}

def topology_key_unrooted(tree):
    tree.is_rooted = False
    tree.encode_bipartitions()
    full_mask = tree.taxon_namespace.all_taxa_bitmask()
    splits = []
    for b in tree.bipartition_encoding:
        if b.is_trivial():
            continue
        x = b.split_bitmask
        x = min(x, full_mask ^ x)
        splits.append(x)
    return tuple(sorted(splits))

# ---------- load ----------
taxa = dendropy.TaxonNamespace()

mb = dendropy.TreeList.get(
    path=trprobs_path,
    schema="nexus",
    taxon_namespace=taxa,
    preserve_underscores=True,
)

# If sampled is nexus, change schema="nexus"
sampled = dendropy.TreeList.get(
    path=sampled_path,
    schema="newick",
    taxon_namespace=taxa,
    preserve_underscores=True,
)

for t in mb:
    t.is_rooted = False
    relabel_tree_inplace(t)

for t in sampled:
    t.is_rooted = False
    relabel_tree_inplace(t)

# ---------- sanity: taxa ----------
mb_leaves = leaf_set(mb[0])
sp_leaves = leaf_set(sampled[0])
shared = mb_leaves & sp_leaves

print("MB leaves:", len(mb_leaves))
print("Sampled leaves:", len(sp_leaves))
print("Shared leaves:", len(shared))
print("Only MB:", sorted(mb_leaves - sp_leaves)[:10], "..." if len(mb_leaves - sp_leaves) > 10 else "")
print("Only sampled:", sorted(sp_leaves - mb_leaves)[:10], "..." if len(sp_leaves - mb_leaves) > 10 else "")

if mb_leaves != sp_leaves:
    raise ValueError("Leaf sets differ. RF/KL comparisons are not valid until labels match exactly.")

# ---------- exact topology overlap ----------
mb_keys = Counter(topology_key_unrooted(t) for t in mb)
sp_keys = Counter(topology_key_unrooted(t) for t in sampled)
overlap = set(mb_keys) & set(sp_keys)

print("\nMB unique topologies:", len(mb_keys))
print("Sampled unique topologies:", len(sp_keys))
print("Exact topology overlap count:", len(overlap))

# ---------- nearest-neighbor RF ----------
# Encode once for speed
for t in mb:
    t.encode_bipartitions()
for t in sampled:
    t.encode_bipartitions()

rf_min = []
argmin_idx = []
for i, s in enumerate(sampled):
    best = None
    best_j = None
    for j, m in enumerate(mb):
        d = treecompare.symmetric_difference(s, m)
        if best is None or d < best:
            best = d
            best_j = j
    rf_min.append(best)
    argmin_idx.append(best_j)

rf_min = np.array(rf_min, dtype=float)
n_taxa = len(mb_leaves)
max_rf = 2 * (n_taxa - 3)

print("\nNum sampled trees:", len(sampled))
print("RF min-to-MB mean:", rf_min.mean())
print("RF min-to-MB median:", np.median(rf_min))
print("RF min-to-MB min/max:", rf_min.min(), rf_min.max())
print("Normalized RF mean:", rf_min.mean() / max_rf)

# ---------- optional histogram ----------
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.hist(rf_min, bins=int(rf_min.max()-rf_min.min()+1), edgecolor="black")
plt.xlabel("min RF to MB set")
plt.ylabel("count")
plt.title("Nearest-neighbor RF from sampled trees to MrBayes trees")
plt.tight_layout()
plt.show()

# ---------- optional: print a few closest trees ----------
closest_idx = np.argsort(rf_min)[:10]
print("\nTop 10 closest sampled trees:")
for rank, i in enumerate(closest_idx, 1):
    print(f"{rank:2d}. sampled_idx={i}, min_RF={rf_min[i]:.0f}, matched_MB_idx={argmin_idx[i]}")
```

```python
from Bio import SeqIO

input_file = "data/seq_gen_1000_toy_data_5_reordered.fasta"
output_file = "data/seq_gen_1000_toy_data_5_reordered.nexus"

# 1. Parse the FASTA file
records = list(SeqIO.parse(input_file, "fasta"))

# 2. Tell Biopython that these sequences are DNA
for record in records:
    record.annotations["molecule_type"] = "DNA"

# 3. Write them to the NEXUS file
count = SeqIO.write(records, output_file, "nexus")
print(f"Converted {count} records to NEXUS format.")

# 4. Append the MrBayes command block automatically
mrbayes_block = """
BEGIN MRBAYES;
    [Set Jukes-Cantor model]
    lset nst=1 basefreq=equal;
    
    [Run MCMC]
    mcmc ngen=100000 samplefreq=100 printfreq=1000 diagnfreq=10000 file=mb_posterior;
    
    [Summarize trees and parameters, discarding 25% burn-in]
    sumt burnin=250;
    sump burnin=250;
    
    quit;
END;
"""

with open(output_file, "a") as f:
    f.write(mrbayes_block)

print("Successfully appended MrBayes execution block!")
```
