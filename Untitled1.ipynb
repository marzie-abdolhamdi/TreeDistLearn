```python
import math
from collections import Counter, defaultdict
import dendropy

# ---------- paths ----------
trprobs_path = "data/DS1/DS1_for_mb.nex.run2.t"  # MrBayes ground-truth distribution
sampled_path = "results_ca_vi/run_20260225_145646_ntaxa27/sampled_trees.nex"  # your sampled trees

# ---------- helpers ----------
def canon_label(x: str) -> str:
    return x.strip().strip("'\"").replace(" ", "_")

def relabel_tree_inplace(tree):
    for nd in tree.leaf_node_iter():
        if nd.taxon is not None and nd.taxon.label is not None:
            nd.taxon.label = canon_label(nd.taxon.label)

def topology_key_unrooted(tree):
    tree.is_rooted = False
    tree.encode_bipartitions()
    full_mask = tree.taxon_namespace.all_taxa_bitmask()
    splits = []
    for b in tree.bipartition_encoding:
        if b.is_trivial():
            continue
        x = b.split_bitmask
        x = min(x, full_mask ^ x)   # canonical orientation
        splits.append(x)
    return tuple(sorted(splits))

def normalize_dict(d):
    z = sum(d.values())
    return {k: v / z for k, v in d.items()} if z > 0 else {}

# ---------- load ----------
taxa = dendropy.TaxonNamespace()

gt_trees = dendropy.TreeList.get(
    path=trprobs_path,
    schema="nexus",
    taxon_namespace=taxa,
    preserve_underscores=True,
)
model_trees = dendropy.TreeList.get(
    path=sampled_path,
    schema="nexus",
    taxon_namespace=taxa,
    preserve_underscores=True,
)

for t in gt_trees:
    relabel_tree_inplace(t)
for t in model_trees:
    relabel_tree_inplace(t)

# ---------- P_GT from trprobs weights ----------
# uses tree.weight if available; fallback weight=1.0
p_counts = defaultdict(float)
for t in gt_trees:
    w = float(t.weight) if getattr(t, "weight", None) is not None else 1.0
    p_counts[topology_key_unrooted(t)] += w
P_gt = normalize_dict(p_counts)

# ---------- Q_model from sampled frequencies ----------
q_counts = Counter(topology_key_unrooted(t) for t in model_trees)
total_q = sum(q_counts.values())

# ---------- KL(P_GT || Q_model) with smoothing on GT support ----------
eps = 1e-10
denom = total_q + eps * len(P_gt)  # smoothing only over GT support
kl = 0.0
for topo, p in P_gt.items():
    q = (q_counts.get(topo, 0) + eps) / denom
    kl += p * (math.log(p) - math.log(q))

# ---------- diagnostics ----------
support_overlap = sum(1 for topo in P_gt if topo in q_counts)
overlap_mass = sum(P_gt[topo] for topo in P_gt if topo in q_counts)

print(f"GT unique topologies: {len(P_gt)}")
print(f"Model unique topologies: {len(q_counts)}")
print(f"Support overlap count: {support_overlap}")
print(f"P_GT mass on overlapping support: {overlap_mass:.6f}")
print(f"KL(P_GT || Q_model): {kl:.10f}")
```

```python
import dendropy
import numpy as np
from scipy.stats import entropy
import matplotlib.pyplot as plt

def compare_trees(mrbayes_path, my_model_path, burnin_fraction=0.25):
    print("Loading trees...")
    
    # 1. Create a shared Taxon Namespace (CRITICAL STEP)
    # This ensures "Taxon_A" is perfectly aligned in memory for both tree sets
    tns = dendropy.TaxonNamespace()
    
    # 2. Load MrBayes Trees
    # MrBayes outputs .t files. We must skip the "burn-in" phase where MCMC was just warming up.
    mb_trees = dendropy.TreeList.get(
        path=mrbayes_path, 
        schema="nexus", 
        taxon_namespace=tns
    )
    n_burnin = int(len(mb_trees) * burnin_fraction)
    mb_trees = mb_trees[n_burnin:] # Keep only the post-burnin trees
    print(f"Loaded {len(mb_trees)} MrBayes trees (after 25% burn-in removal).")

    # 3. Load Your Model's Trees
    my_trees = dendropy.TreeList.get(
        path=my_model_path, 
        schema="nexus", 
        taxon_namespace=tns
    )
    print(f"Loaded {len(my_trees)} Variational Inference trees.")

    # 4. Helper function to safely calculate split frequencies
    def get_split_frequencies(tree_list):
        freq_dict = {}
        n_trees = len(tree_list)
        
        for tree in tree_list:
            # Tell DendroPy to treat the trees as unrooted (standard for this math)
            tree.is_rooted = False 
            # Encode the bipartition on the individual tree
            tree.encode_bipartitions()
            
            for bip in tree.bipartition_encoding:
                # We use the split bitmask (an integer) as the key to guarantee perfect matching
                split_key = bip.split_bitmask
                freq_dict[split_key] = freq_dict.get(split_key, 0) + 1
                
        # Convert raw counts to probabilities between 0.0 and 1.0
        for key in freq_dict:
            freq_dict[key] /= n_trees
            
        return freq_dict

    print("Encoding bipartitions...")
    mb_freqs = get_split_frequencies(mb_trees)
    my_freqs = get_split_frequencies(my_trees)

    # 5. Align the frequencies for comparison
    # We find every unique split that exists in EITHER model
    all_bipartitions = set(mb_freqs.keys()).union(set(my_freqs.keys()))

    p_mb = [] # Ground Truth (MrBayes)
    q_my = [] # Estimated (Your Model)

    for bp in all_bipartitions:
        p_mb.append(mb_freqs.get(bp, 0.0))
        q_my.append(my_freqs.get(bp, 0.0))

    p_mb = np.array(p_mb)
    q_my = np.array(q_my)
    # --- METRICS CALCULATION ---
    
    # Metric 1: Mean Absolute Error (MAE)
    mae = np.mean(np.abs(p_mb - q_my))
    print(f"\nMean Absolute Error (MAE) of splits: {mae:.4f}")
    
    # Metric 2: KL Divergence
    # We add a tiny epsilon (1e-10) so we don't calculate log(0) if a split is missing
    epsilon = 1e-10
    p_mb_smooth = p_mb + epsilon
    q_my_smooth = q_my + epsilon
    
    # Normalize to create valid probability distributions
    p_mb_smooth /= np.sum(p_mb_smooth)
    q_my_smooth /= np.sum(q_my_smooth)
    
    kl_div = entropy(p_mb_smooth, q_my_smooth)
    print(f"KL Divergence (Splits): {kl_div:.4f}")
    
    # Metric 3: Pearson Correlation
    correlation = np.corrcoef(p_mb, q_my)[0, 1]
    print(f"Pearson Correlation (r): {correlation:.4f}")

    # --- PLOTTING FOR YOUR PRESENTATION ---
    plt.figure(figsize=(6, 6))
    plt.scatter(p_mb, q_my, alpha=0.5, color='blue', edgecolor='k')
    
    # Draw the perfect diagonal line
    plt.plot([0, 1], [0, 1], color='red', linestyle='--', label="Perfect Match")
    
    plt.title("Split Frequencies: MrBayes vs Variational Inference")
    plt.xlabel("MrBayes True Posterior Probability")
    plt.ylabel("Learned VI Posterior Probability")
    plt.xlim(-0.05, 1.05)
    plt.ylim(-0.05, 1.05)
    plt.grid(True, alpha=0.3)
    plt.legend()
    
    plt.tight_layout()
    plt.savefig("posterior_comparison.png", dpi=300)
    print("\nSaved scatter plot to 'posterior_comparison.png'")

if __name__ == "__main__":
    # Replace these with your actual file names!
    compare_trees(
        mrbayes_path="data/DS1_synth_1000/seq_gen_DS1_for_mb.nex.run1.t", 
        my_model_path="results_ca_vi/run_20260226_170508_ntaxa27/sampled_trees.nex"
    )
```

```python
import dendropy
import numpy as np
from collections import Counter

def get_topology_frequencies(tree_file, schema="nexus"):
    """
    Reads a file of trees, extracts the unique unrooted topologies 
    (ignoring branch lengths), and calculates their frequencies.
    """
    print(f"Loading trees from {tree_file}...")
    
    # Load all trees into a single namespace so the splits align perfectly
    tns = dendropy.TaxonNamespace()
    trees = dendropy.TreeList.get(path=tree_file, schema=schema, taxon_namespace=tns)
    
    topology_counts = Counter()
    
    for tree in trees:
        # Standardize the tree to unrooted topology
        tree.is_rooted = False
        
        # FIX: Use the modern DendroPy 4.x method to remove degree-2 nodes
        tree.suppress_unifurcations() 
        
        # Encode the tree as a set of mathematical bipartitions (splits)
        tree.encode_bipartitions()
        
        # Create a unique, sortable tuple representing this exact topology
        splits = tuple(sorted([split.split_as_int() for split in tree.bipartition_encoding]))
        topology_counts[splits] += 1
        
    total_trees = len(trees)
    print(f"Found {total_trees} total trees, containing {len(topology_counts)} unique topologies.")
    
    return topology_counts, total_trees, tns

def calculate_empirical_kl(mrbayes_file, model_file, mrbayes_schema="nexus", model_schema="nexus", epsilon=1e-6):
    """
    Calculates the KL divergence: sum( P(t) * log( P(t) / Q(t) ) )
    P = MrBayes (Ground Truth)
    Q = Model Samples
    """
    # 1. Get the counts for both distributions
    mb_counts, mb_total, tns = get_topology_frequencies(mrbayes_file, schema=mrbayes_schema)
    
    # Load model trees using the exact same taxon namespace (tns)
    mod_trees = dendropy.TreeList.get(path=model_file, schema=model_schema, taxon_namespace=tns)
    mod_counts = Counter()
    for tree in mod_trees:
        tree.is_rooted = False
        
        # FIX: Use the modern DendroPy 4.x method here too
        tree.suppress_unifurcations()
        
        tree.encode_bipartitions()
        splits = tuple(sorted([split.split_as_int() for split in tree.bipartition_encoding]))
        mod_counts[splits] += 1
    mod_total = len(mod_trees)

    # 2. Calculate KL Divergence
    kl_div = 0.0
    missed_topologies = 0
    
    for topo, count in mb_counts.items():
        # P(tau) - Ground truth probability
        p_tau = count / mb_total
        
        # Q(tau) - Model probability
        if topo in mod_counts:
            q_tau = mod_counts[topo] / mod_total
            # Smooth slightly to prevent domain errors
            q_tau = (q_tau + epsilon) / (1.0 + epsilon)
        else:
            q_tau = epsilon / (1.0 + epsilon)
            missed_topologies += 1
            
        kl_div += p_tau * np.log(p_tau / q_tau)
        
    print(f"\n--- Results ---")
    print(f"Model missed {missed_topologies} out of {len(mb_counts)} unique MrBayes topologies.")
    print(f"KL Divergence: {kl_div:.4f}")
    
    return kl_div

if __name__ == "__main__":
    # Point these to your specific files!
    MRBAYES_POSTERIOR_FILE = "data/DS1.trprobs" 
    MODEL_SAMPLED_FILE = "results_converge_elbo/run_20260218_172751_ntaxa27/sampled_trees_standalone.nex"
    
    calculate_empirical_kl(
        mrbayes_file=MRBAYES_POSTERIOR_FILE, 
        model_file=MODEL_SAMPLED_FILE,
        mrbayes_schema="nexus", 
        model_schema="nexus"    
    )
```

```python
# --- Sparsity pattern (gaps) plot for an alignment FASTA (.fasta / .fa) ---

from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt


# ---------------------------
# 1) FASTA alignment reader
# ---------------------------
def read_fasta_alignment(fasta_path: str):
    """
    Reads a FASTA multiple sequence alignment.
    Returns:
        names: list[str]
        seqs:  list[str] (all same length)
    """
    fasta_path = Path(fasta_path)
    if not fasta_path.exists():
        raise FileNotFoundError(f"File not found: {fasta_path}")

    names, seqs = [], []
    cur_name, cur_seq = None, []

    with fasta_path.open("r", encoding="utf-8", errors="ignore") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            if line.startswith(">"):
                # flush previous
                if cur_name is not None:
                    names.append(cur_name)
                    seqs.append("".join(cur_seq))
                cur_name = line[1:].strip()
                cur_seq = []
            else:
                cur_seq.append(line)

        # flush last
        if cur_name is not None:
            names.append(cur_name)
            seqs.append("".join(cur_seq))

    if not seqs:
        raise ValueError("No sequences found. Is this a valid FASTA?")

    Ls = {len(s) for s in seqs}
    if len(Ls) != 1:
        # If this triggers, your FASTA is not an aligned MSA (or has line issues).
        raise ValueError(f"Sequences have different lengths: {sorted(Ls)}. "
                         "This does not look like an alignment FASTA.")

    return names, seqs


# ---------------------------
# 2) Convert to gap mask
# ---------------------------
def alignment_gap_mask(seqs, gap_chars=("-", "."), uppercase=True):
    """
    Converts sequences -> boolean mask (N x L):
        True  = gap at [i, j]
        False = not-gap
    """
    if uppercase:
        seqs = [s.upper() for s in seqs]

    N = len(seqs)
    L = len(seqs[0])
    # Create a (N,L) array of characters
    arr = np.frombuffer("".join(seqs).encode("ascii", "ignore"), dtype="S1").reshape(N, L)
    # Mask gaps
    gap_bytes = {c.encode("ascii") for c in gap_chars}
    mask = np.isin(arr, list(gap_bytes))
    return mask


# ---------------------------
# 3) Plot sparsity pattern
# ---------------------------
def plot_gap_sparsity(mask, title="Gap sparsity pattern", max_taxa=None, max_sites=None):
    """
    mask: boolean (N x L), True = gap
    max_taxa/max_sites: optional downsampling (for huge alignments).
    """
    N, L = mask.shape
    mask_plot = mask

    if max_taxa is not None and N > max_taxa:
        # evenly sample rows
        idx = np.linspace(0, N - 1, max_taxa).round().astype(int)
        mask_plot = mask_plot[idx, :]
        N = mask_plot.shape[0]

    if max_sites is not None and L > max_sites:
        # evenly sample columns
        jdx = np.linspace(0, L - 1, max_sites).round().astype(int)
        mask_plot = mask_plot[:, jdx]
        L = mask_plot.shape[1]

    plt.figure(figsize=(12, 6))
    # show True as black pixels, False as white
    plt.imshow(mask_plot, aspect="auto", interpolation="nearest", cmap="gray_r")
    plt.xlabel("Alignment site (column)")
    plt.ylabel("Taxon (sequence index)")
    plt.title(f"{title}\n(black = gap, white = residue) | N={mask.shape[0]}, L={mask.shape[1]}")
    plt.tight_layout()
    plt.show()


# ---------------------------
# 4) Useful summary plots
# ---------------------------
def plot_gap_summaries(mask):
    """
    Adds:
      - gap fraction per site (column)
      - gap fraction per taxon (row)
    """
    gap_per_site = mask.mean(axis=0)  # length L
    gap_per_taxon = mask.mean(axis=1) # length N

    plt.figure(figsize=(12, 3))
    plt.plot(gap_per_site)
    plt.xlabel("Alignment site (column)")
    plt.ylabel("Gap fraction")
    plt.title("Gap fraction per site")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(12, 3))
    plt.plot(gap_per_taxon)
    plt.xlabel("Taxon (sequence index)")
    plt.ylabel("Gap fraction")
    plt.title("Gap fraction per taxon")
    plt.tight_layout()
    plt.show()


# ---------------------------
# 5) Run it on your file
# ---------------------------
FASTA_PATH = "data/DS1_reordered_real.fasta"  # <-- change this to your path

names, seqs = read_fasta_alignment(FASTA_PATH)
mask = alignment_gap_mask(seqs, gap_chars=("-", "."))  # treat '-' and '.' as gaps

print(f"Loaded alignment: N={len(seqs)} taxa, L={len(seqs[0])} sites")
print(f"Overall gap fraction: {mask.mean():.4f}")

plot_gap_sparsity(mask, title=f"Gap sparsity: {Path(FASTA_PATH).name}")
plot_gap_summaries(mask)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_gap_sparsity_highlight_all_gap_cols(
    mask,
    title="Gap sparsity pattern",
    max_taxa=None,
    max_sites=None,
    highlight_color="red",
    highlight_alpha=0.35,
):
    """
    mask: boolean (N x L), True = gap
    Highlights columns where ALL taxa are gaps.
    """
    N0, L0 = mask.shape
    mask_plot = mask

    # Optional downsampling (keep mapping indices so we still detect all-gap columns correctly)
    row_idx = np.arange(N0)
    col_idx = np.arange(L0)

    if max_taxa is not None and N0 > max_taxa:
        row_idx = np.linspace(0, N0 - 1, max_taxa).round().astype(int)
        mask_plot = mask_plot[row_idx, :]

    if max_sites is not None and L0 > max_sites:
        col_idx = np.linspace(0, L0 - 1, max_sites).round().astype(int)
        mask_plot = mask_plot[:, col_idx]

    # Detect "all-gap" columns in the plotted mask
    # (all taxa in that plotted view are gaps)
    all_gap_cols = mask_plot.all(axis=0)  # length L_plot

    # Base plot: black for gaps, white for residues
    plt.figure(figsize=(12, 6))
    plt.imshow(mask_plot, aspect="auto", interpolation="nearest", cmap="gray_r")
    plt.xlabel("Alignment site (column)")
    plt.ylabel("Taxon (sequence index)")
    plt.title(
        f"{title}\n(black = gap, white = residue) | N={N0}, L={L0} | "
        f"all-gap cols shown in {highlight_color}"
    )

    # Overlay highlight bands for all-gap columns
    # Each band spans x in [j-0.5, j+0.5]
    ys = (-0.5, mask_plot.shape[0] - 0.5)
    for j in np.where(all_gap_cols)[0]:
        plt.fill_betweenx(ys, j - 0.5, j + 0.5, color=highlight_color, alpha=highlight_alpha, linewidth=0)

    plt.tight_layout()
    plt.show()

    # Print which columns are all-gap (in original coordinates if downsampled)
    all_gap_cols_orig = col_idx[np.where(all_gap_cols)[0]]
    print(f"All-gap columns (count={len(all_gap_cols_orig)}):")
    print(all_gap_cols_orig[:200], "..." if len(all_gap_cols_orig) > 200 else "")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_gap_sparsity_highlight_sparse_cols(
    mask,
    threshold=0.80,          # gap fraction threshold (0..1)
    title="Gap sparsity pattern",
    max_taxa=None,
    max_sites=None,
    highlight_color="orange",
    highlight_alpha=0.30,
):
    """
    mask: boolean (N x L), True = gap
    Highlights columns where (gap fraction) >= threshold.
    """
    N0, L0 = mask.shape
    mask_plot = mask

    # Optional downsampling (keep mapping indices for reporting)
    row_idx = np.arange(N0)
    col_idx = np.arange(L0)

    if max_taxa is not None and N0 > max_taxa:
        row_idx = np.linspace(0, N0 - 1, max_taxa).round().astype(int)
        mask_plot = mask_plot[row_idx, :]

    if max_sites is not None and L0 > max_sites:
        col_idx = np.linspace(0, L0 - 1, max_sites).round().astype(int)
        mask_plot = mask_plot[:, col_idx]

    # Gap fraction per column in the plotted view
    gap_frac = mask_plot.mean(axis=0)  # length L_plot
    sparse_cols = gap_frac >= threshold

    # Base plot: black gaps, white residues
    plt.figure(figsize=(12, 6))
    plt.imshow(mask_plot, aspect="auto", interpolation="nearest", cmap="gray_r")
    plt.xlabel("Alignment site (column)")
    plt.ylabel("Taxon (sequence index)")
    plt.title(
        f"{title}\n(black = gap, white = residue) | N={N0}, L={L0} | "
        f"highlight: gap_fraction ≥ {threshold:.0%}"
    )

    # Overlay highlight bands
    ys = (-0.5, mask_plot.shape[0] - 0.5)
    for j in np.where(sparse_cols)[0]:
        plt.fill_betweenx(ys, j - 0.5, j + 0.5, color=highlight_color,
                          alpha=highlight_alpha, linewidth=0)

    plt.tight_layout()
    plt.show()

    # Report in original coordinates (if downsampled)
    sparse_cols_orig = col_idx[np.where(sparse_cols)[0]]
    print(f"Sparse columns with gap_fraction ≥ {threshold:.0%}: {len(sparse_cols_orig)} / {L0}")
    if len(sparse_cols_orig) > 0:
        print("First few sparse column indices:", sparse_cols_orig[:50])
        # also show summary stats
        frac_orig = mask[:, sparse_cols_orig].mean(axis=0) if len(sparse_cols_orig) else np.array([])
        if len(frac_orig):
            print(f"Gap fraction among highlighted columns: min={frac_orig.min():.3f}, "
                  f"median={np.median(frac_orig):.3f}, max={frac_orig.max():.3f}")


def plot_gap_fraction_per_site(mask, threshold=0.80):
    """Simple 1D plot of gap fraction per column + threshold line."""
    gap_per_site = mask.mean(axis=0)
    plt.figure(figsize=(12, 3))
    plt.plot(gap_per_site)
    plt.axhline(threshold, linestyle="--")
    plt.xlabel("Alignment site (column)")
    plt.ylabel("Gap fraction")
    plt.title(f"Gap fraction per site (threshold = {threshold:.0%})")
    plt.tight_layout()
    plt.show()
```

```python
# Choose your threshold here (e.g. 0.80 for 80%)
thr = 0.80

plot_gap_sparsity_highlight_sparse_cols(
    mask,
    threshold=thr,
    title="Gap sparsity with sparse columns highlighted",
    highlight_color="red",
    highlight_alpha=0.25,
    # max_taxa=500, max_sites=5000  # optional for big MSAs
)

# Optional: also plot the per-site gap fraction curve
plot_gap_fraction_per_site(mask, threshold=thr)
```
