```python
# Notebook cell: plot edge embeddings h_e (post-training) for one insertion step
# Works with your existing run folder + saved edge_posterior_net.pt files.

import os
import re
import json
import importlib.util
import numpy as np
import matplotlib.pyplot as plt
import torch
from sklearn.decomposition import PCA

# -------------------------
# USER INPUTS
# -------------------------
script_path = "taxa_inseter_final_best_2_blopt.py"
run_dir = "results_full/run_20260209_104253_ntaxa27"   # your run folder
fasta_path = "data/seq_gen_1000_DS1.fasta"    # same fasta used in training
step_taxon_idx = 20                                     # which training step to visualize

# -------------------------
# load your script as module
# -------------------------
spec = importlib.util.spec_from_file_location("taxa_mod", script_path)
taxa_mod = importlib.util.module_from_spec(spec)
spec.loader.exec_module(taxa_mod)

# -------------------------
# load run metadata
# -------------------------
with open(os.path.join(run_dir, "config.json"), "r", encoding="utf-8") as f:
    cfg = json.load(f)

with open(os.path.join(run_dir, "step_meta.json"), "r", encoding="utf-8") as f:
    step_meta = json.load(f)

if cfg.get("bl_opt_after_each_insert", False):
    print("Warning: --bl_opt_after_each_insert was enabled. Reconstructed tree state may not exactly match training-time lengths unless you saved per-step trees.")

# -------------------------
# build leaf likelihoods
# -------------------------
names_all, seqs_all = taxa_mod.parse_fasta_in_order(fasta_path)
n_use = min(cfg.get("n_taxa_use", len(names_all)) or len(names_all), len(names_all))
names = names_all[:n_use]
seqs = seqs_all[:n_use]
leaf_lik_all = {i: taxa_mod.seq_to_likelihood_matrix(seqs[i]) for i in range(n_use)}

# -------------------------
# rebuild tree state BEFORE step_taxon_idx insertion
# -------------------------
allocator = taxa_mod.InternalNodeAllocator(start=n_use)
center_id = allocator.alloc()
tree = taxa_mod.build_t3_star(cfg["edge_len_0"], cfg["edge_len_1"], cfg["edge_len_2"], center_idx=center_id)

# replay insertions up to step_taxon_idx-1
for rec in step_meta:
    tidx = int(rec["taxon_idx"])
    if tidx >= step_taxon_idx:
        break
    edge = tuple(rec["chosen_edge"])
    rho = float(rec["chosen_rho"])
    b = float(rec["chosen_b"])
    taxa_mod.insert_taxon_on_edge(tree, edge=edge, new_taxon=tidx, rho=rho, b_len=b, allocator=allocator)

# -------------------------
# load net for this step
# -------------------------
step_dir = None
for d in os.listdir(run_dir):
    m = re.match(rf"^train_step_{step_taxon_idx:03d}_", d)
    if m:
        step_dir = os.path.join(run_dir, d)
        break

if step_dir is None:
    raise FileNotFoundError(f"Could not find train_step directory for taxon_idx={step_taxon_idx}")

model_path = os.path.join(step_dir, "edge_posterior_net.pt")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# edge features for current tree
leaf_lik_current = {k: v for k, v in leaf_lik_all.items() if k in tree.adj and len(tree.adj[k]) == 1}
msg_cache, _ = taxa_mod.compute_messages_and_cavities(tree, leaf_lik_current)
edges = tree.edges()
edge_feats = torch.stack(
    [taxa_mod.edge_feature(msg_cache[(u, v)], msg_cache[(v, u)]) for (u, v) in edges], dim=0
).to(device)

# instantiate and load net
hidden_dim = int(cfg["hidden_dim"])
net = taxa_mod.EdgePosteriorNet(in_dim=16, hidden=hidden_dim).to(device)
net.load_state_dict(torch.load(model_path, map_location=device))
net.eval()

with torch.no_grad():
    h = net.backbone(edge_feats)  # <- edge embedding h_e
    edge_logit, *_ = net(edge_feats)
    q = torch.softmax(edge_logit, dim=0)

H = h.cpu().numpy()
Q = q.cpu().numpy()

# PCA to 2D
pca = PCA(n_components=2, random_state=0)
Z = pca.fit_transform(H)

# labels
def node_label(node_id):
    return names[node_id] if node_id < len(names) else f"i{node_id}"

edge_labels = [f"({node_label(u)}, {node_label(v)})" for (u, v) in edges]

# plot
plt.figure(figsize=(8, 6))
sc = plt.scatter(Z[:, 0], Z[:, 1], c=Q, s=40 + 500 * Q, cmap="viridis")
plt.colorbar(sc, label="q(e)")
plt.title(f"Edge Embeddings h_e (PCA) at step taxon_idx={step_taxon_idx}")
plt.xlabel("PC1")
plt.ylabel("PC2")

# annotate top edges
topk = np.argsort(Q)[-5:]
for i in topk:
    plt.annotate(edge_labels[i], (Z[i, 0], Z[i, 1]), fontsize=8)

plt.tight_layout()
plt.show()

print("Top edges by q(e):")
for i in topk[::-1]:
    print(f"{edge_labels[i]}   q={Q[i]:.4f}")
```

```python
# Notebook cell 2: loop all steps, save one PCA embedding figure per step
# Requires the variables and module from cell 1:
# - taxa_mod, run_dir, cfg, step_meta, names, leaf_lik_all

import os
import numpy as np
import matplotlib.pyplot as plt
import torch
from sklearn.decomposition import PCA

out_dir = os.path.join(run_dir, "edge_embedding_pca_all_steps")
os.makedirs(out_dir, exist_ok=True)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
hidden_dim = int(cfg["hidden_dim"])

# helper: labels
def node_label(node_id):
    return names[node_id] if node_id < len(names) else f"i{node_id}"

# rebuild tree from scratch, then advance step-by-step
n_use = len(names)
allocator = taxa_mod.InternalNodeAllocator(start=n_use)
center_id = allocator.alloc()
tree = taxa_mod.build_t3_star(cfg["edge_len_0"], cfg["edge_len_1"], cfg["edge_len_2"], center_idx=center_id)

# index step_meta by taxon_idx for quick access
meta_by_idx = {int(r["taxon_idx"]): r for r in step_meta}
step_indices = sorted(meta_by_idx.keys())

for sidx in step_indices:
    # tree is currently state BEFORE inserting sidx (if we maintain below)
    rec = meta_by_idx[sidx]

    # find model file
    step_dir = None
    for d in os.listdir(run_dir):
        if d.startswith(f"train_step_{sidx:03d}_"):
            step_dir = os.path.join(run_dir, d)
            break
    if step_dir is None:
        print(f"[skip] step {sidx}: train_step dir not found")
        # still perform insertion replay so next steps are reachable
        taxa_mod.insert_taxon_on_edge(
            tree=tree,
            edge=tuple(rec["chosen_edge"]),
            new_taxon=sidx,
            rho=float(rec["chosen_rho"]),
            b_len=float(rec["chosen_b"]),
            allocator=allocator,
        )
        continue

    model_path = os.path.join(step_dir, "edge_posterior_net.pt")
    if not os.path.exists(model_path):
        print(f"[skip] step {sidx}: model not found")
        taxa_mod.insert_taxon_on_edge(
            tree=tree,
            edge=tuple(rec["chosen_edge"]),
            new_taxon=sidx,
            rho=float(rec["chosen_rho"]),
            b_len=float(rec["chosen_b"]),
            allocator=allocator,
        )
        continue

    # build edge features for current tree state
    leaf_lik_current = {k: v for k, v in leaf_lik_all.items() if k in tree.adj and len(tree.adj[k]) == 1}
    msg_cache, _ = taxa_mod.compute_messages_and_cavities(tree, leaf_lik_current)
    edges = tree.edges()
    edge_feats = torch.stack(
        [taxa_mod.edge_feature(msg_cache[(u, v)], msg_cache[(v, u)]) for (u, v) in edges], dim=0
    ).to(device)

    # load net and get embeddings
    net = taxa_mod.EdgePosteriorNet(in_dim=16, hidden=hidden_dim).to(device)
    net.load_state_dict(torch.load(model_path, map_location=device))
    net.eval()

    with torch.no_grad():
        H = net.backbone(edge_feats).cpu().numpy()
        edge_logit, *_ = net(edge_feats)
        Q = torch.softmax(edge_logit, dim=0).cpu().numpy()

    # PCA
    if H.shape[0] >= 2 and H.shape[1] >= 2:
        pca = PCA(n_components=2, random_state=0)
        Z = pca.fit_transform(H)
    else:
        Z = np.zeros((H.shape[0], 2), dtype=np.float32)

    edge_labels = [f"({node_label(u)}, {node_label(v)})" for (u, v) in edges]
    topk = np.argsort(Q)[-min(5, len(Q)):]

    # plot
    plt.figure(figsize=(8, 6))
    sc = plt.scatter(Z[:, 0], Z[:, 1], c=Q, s=40 + 500 * Q, cmap="viridis")
    plt.colorbar(sc, label="q(e)")
    plt.title(f"Edge Embeddings h_e (PCA) | step taxon_idx={sidx}")
    plt.xlabel("PC1")
    plt.ylabel("PC2")

    for i in topk:
        plt.annotate(edge_labels[i], (Z[i, 0], Z[i, 1]), fontsize=7)

    plt.tight_layout()
    fig_path = os.path.join(out_dir, f"step_{sidx:03d}_edge_embedding_pca.png")
    plt.savefig(fig_path, dpi=180)
    plt.close()

    # also save top edges table
    txt_path = os.path.join(out_dir, f"step_{sidx:03d}_top_edges.txt")
    with open(txt_path, "w", encoding="utf-8") as f:
        f.write(f"step_taxon_idx={sidx}\n")
        for i in topk[::-1]:
            f.write(f"{edge_labels[i]}\tq={Q[i]:.8f}\n")

    # replay insertion so next loop sees next-step tree
    taxa_mod.insert_taxon_on_edge(
        tree=tree,
        edge=tuple(rec["chosen_edge"]),
        new_taxon=sidx,
        rho=float(rec["chosen_rho"]),
        b_len=float(rec["chosen_b"]),
        allocator=allocator,
    )

print(f"Saved all PCA figures to: {out_dir}")
```

```python
# Companion diagnostics for one step:
# 1) q(e) bar plot
# 2) expected log-likelihood per edge (and relative version)

import os, re, json, importlib.util
import numpy as np
import matplotlib.pyplot as plt
import torch
from argparse import Namespace

# ---------- USER INPUT ----------
script_path = "taxa_inseter_final_best_2_blopt.py"
run_dir = "results_full/run_20260209_104253_ntaxa27"   # your run folder
fasta_path = "data/seq_gen_1000_DS1.fasta"    # same fasta used in training
step_taxon_idx = 20                                     # which training step to visualize

k_eval = 400  # MC samples for E_q[log p] estimate (increase for smoother bars)

# ---------- load module ----------
spec = importlib.util.spec_from_file_location("taxa_mod", script_path)
taxa_mod = importlib.util.module_from_spec(spec)
spec.loader.exec_module(taxa_mod)

# ---------- load run metadata ----------
with open(os.path.join(run_dir, "config.json"), "r", encoding="utf-8") as f:
    cfg = json.load(f)
with open(os.path.join(run_dir, "step_meta.json"), "r", encoding="utf-8") as f:
    step_meta = json.load(f)

# ---------- build leaf likelihoods ----------
names_all, seqs_all = taxa_mod.parse_fasta_in_order(fasta_path)
n_use = min(cfg.get("n_taxa_use", len(names_all)) or len(names_all), len(names_all))
names = names_all[:n_use]
seqs = seqs_all[:n_use]
leaf_lik_all = {i: taxa_mod.seq_to_likelihood_matrix(seqs[i]) for i in range(n_use)}

def node_label(node_id):
    return names[node_id] if node_id < len(names) else f"i{node_id}"

# ---------- reconstruct tree BEFORE this step ----------
allocator = taxa_mod.InternalNodeAllocator(start=n_use)
center_id = allocator.alloc()
tree = taxa_mod.build_t3_star(cfg["edge_len_0"], cfg["edge_len_1"], cfg["edge_len_2"], center_idx=center_id)

for rec in step_meta:
    tidx = int(rec["taxon_idx"])
    if tidx >= step_taxon_idx:
        break
    taxa_mod.insert_taxon_on_edge(
        tree=tree,
        edge=tuple(rec["chosen_edge"]),
        new_taxon=tidx,
        rho=float(rec["chosen_rho"]),
        b_len=float(rec["chosen_b"]),
        allocator=allocator,
    )

# ---------- locate model for this step ----------
step_dir = None
for d in os.listdir(run_dir):
    if d.startswith(f"train_step_{step_taxon_idx:03d}_"):
        step_dir = os.path.join(run_dir, d)
        break
if step_dir is None:
    raise FileNotFoundError(f"train_step dir not found for taxon_idx={step_taxon_idx}")

model_path = os.path.join(step_dir, "edge_posterior_net.pt")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ---------- compute messages/features ----------
leaf_lik_current = {k: v for k, v in leaf_lik_all.items() if k in tree.adj and len(tree.adj[k]) == 1}
msg_cache, cav_cache = taxa_mod.compute_messages_and_cavities(tree, leaf_lik_current)
edges = tree.edges()
edge_feats = torch.stack(
    [taxa_mod.edge_feature(msg_cache[(u, v)], msg_cache[(v, u)]) for (u, v) in edges], dim=0
).to(device)

# ---------- load net ----------
net = taxa_mod.EdgePosteriorNet(in_dim=16, hidden=int(cfg["hidden_dim"])).to(device)
net.load_state_dict(torch.load(model_path, map_location=device))
net.eval()

with torch.no_grad():
    edge_logit, mu_b, log_sigma_b, mu_r, log_sigma_r = net(edge_feats)
    q_e = torch.softmax(edge_logit, dim=0)

    # expected log-likelihood under q(rho,b|e)
    loglik_mean, _, _ = taxa_mod.mc_expected_loglik_and_samples(
        tree=tree,
        edges=edges,
        mu_b=mu_b,
        log_sigma_b=log_sigma_b,
        mu_r=mu_r,
        log_sigma_r=log_sigma_r,
        new_leaf_lik=leaf_lik_all[step_taxon_idx].to(device),
        cav_cache=cav_cache,
        pi=taxa_mod.PI_DNA.to(device),
        k_samples=k_eval,
        jc_rate=cfg.get("jc_rate", 1.0),
    )

q_np = q_e.cpu().numpy()
ll_np = loglik_mean.cpu().numpy()
ll_rel = ll_np - ll_np.max()

edge_labels = [f"({node_label(u)}, {node_label(v)})" for (u, v) in edges]
order = np.argsort(q_np)[::-1]  # sort by posterior mass
x = np.arange(len(edges))

# ---------- plots ----------
fig, axes = plt.subplots(2, 1, figsize=(14, 10), sharex=True)

axes[0].bar(x, q_np[order])
axes[0].set_ylabel("q(e)")
axes[0].set_title(f"Step {step_taxon_idx}: Edge Posterior q(e)")

axes[1].bar(x, ll_np[order], alpha=0.7, label="E_q[log p(y|e,rho,b)]")
axes[1].plot(x, ll_rel[order], color="red", linewidth=2, label="relative (minus max)")
axes[1].set_ylabel("Expected log-likelihood")
axes[1].set_title(f"Step {step_taxon_idx}: Edge-conditioned Expected Log-Likelihood")
axes[1].legend()

plt.xticks(x, [edge_labels[i] for i in order], rotation=60, ha="right")
plt.tight_layout()
plt.show()

# ---------- top edges summary ----------
print("Top 10 edges by q(e):")
for rank, i in enumerate(order[:10], start=1):
    print(f"{rank:2d}. {edge_labels[i]:40s}  q={q_np[i]:.6f}  Eloglik={ll_np[i]:.3f}  rel={ll_rel[i]:.3f}")
```

```python
# Same-step overlay: normalized q(e) vs normalized relative expected log-likelihood
# Assumes you already computed: q_np, ll_np, ll_rel, edge_labels, order

import numpy as np
import matplotlib.pyplot as plt

def minmax01(a):
    a = np.asarray(a, dtype=float)
    lo, hi = a.min(), a.max()
    if hi - lo < 1e-12:
        return np.zeros_like(a)
    return (a - lo) / (hi - lo)

# Convert ll to "higher is better" score
ll_score = ll_rel - ll_rel.min()   # now min=0, max=best gap
ll_norm = minmax01(ll_score)
q_norm = minmax01(q_np)

x = np.arange(len(edge_labels))
idx = order  # already sorted by q descending

plt.figure(figsize=(14, 6))
plt.plot(x, q_norm[idx], marker="o", linewidth=2, label="q(e) normalized")
plt.plot(x, ll_norm[idx], marker="s", linewidth=2, label="E_q[log p] normalized (relative)")
plt.xticks(x, [edge_labels[i] for i in idx], rotation=60, ha="right")
plt.ylim(-0.02, 1.05)
plt.ylabel("Normalized score [0,1]")
plt.title("Edge ranking: posterior vs expected log-likelihood (normalized)")
plt.legend()
plt.tight_layout()
plt.show()

# Optional rank agreement summary
rank_q = np.argsort(np.argsort(-q_np))
rank_ll = np.argsort(np.argsort(-ll_score))
rho = np.corrcoef(rank_q, rank_ll)[0, 1]
print(f"Rank correlation (Pearson on rank indices): {rho:.4f}")

topk = 10
print(f"\nTop-{topk} by q(e):")
for r, i in enumerate(idx[:topk], 1):
    print(f"{r:2d}. {edge_labels[i]:40s} q={q_np[i]:.6f} qN={q_norm[i]:.3f} llRel={ll_rel[i]:.3f} llN={ll_norm[i]:.3f}")
```
