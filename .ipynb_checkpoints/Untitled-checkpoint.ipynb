```python
import dendropy
from dendropy.calculate import treecompare

sampled_nexus_path = "results_full/run_20260211_222557_ntaxa27/sampled_trees.nex"
ground_truth_newick_path = "data/ds1_best_mrbayes.nwk"

taxa = dendropy.TaxonNamespace()

sampled = dendropy.TreeList.get(
    path=sampled_nexus_path,
    schema="nexus",
    taxon_namespace=taxa,
    preserve_underscores=True,
)

gt = dendropy.Tree.get(
    path=ground_truth_newick_path,
    schema="newick",
    taxon_namespace=taxa,
    preserve_underscores=True,
)

gt.is_rooted = False
gt.encode_bipartitions()

gt_labels = {nd.taxon.label for nd in gt.leaf_node_iter()}

rfs = []
n_shared_each = []

for t in sampled:
    t.is_rooted = False
    t.encode_bipartitions()

    t_labels = {nd.taxon.label for nd in t.leaf_node_iter()}
    shared = t_labels & gt_labels
    n_shared_each.append(len(shared))

    rf = treecompare.symmetric_difference(t, gt)
    rfs.append(rf)

# Usually all same; take first if consistent
if len(set(n_shared_each)) == 1:
    n_taxa = n_shared_each[0]
else:
    print("Warning: shared taxa count varies across sampled trees:", sorted(set(n_shared_each)))
    n_taxa = min(n_shared_each)

max_rf = 2 * (n_taxa - 3) if n_taxa >= 3 else float("nan")

rf_mean = sum(rfs) / len(rfs)

print(f"Num sampled trees: {len(rfs)}")
print(f"Shared taxa (per tree): min={min(n_shared_each)}, max={max(n_shared_each)}")
print(f"RF mean: {rf_mean:.4f}")
print(f"RF min/max: {min(rfs)} / {max(rfs)}")
print(f"Max RF (n={n_taxa}): {max_rf}")
print(f"Normalized RF mean: {rf_mean/max_rf:.4f}" if max_rf == max_rf and max_rf > 0 else "Normalized RF mean: nan")
```

```python
import itertools
import math
import dendropy
```

```python
def load_tree(path, schema, taxa):
    t = dendropy.Tree.get(path=path, schema=schema, taxon_namespace=taxa)
    t.is_rooted = False
    return t

def quartet_split_id(pdm, ta, tb, tc, td, tol=1e-10):
    # For unrooted additive trees, the smallest of these 3 sums gives the quartet split
    s1 = pdm.distance(ta, tb) + pdm.distance(tc, td)  # ab|cd
    s2 = pdm.distance(ta, tc) + pdm.distance(tb, td)  # ac|bd
    s3 = pdm.distance(ta, td) + pdm.distance(tb, tc)  # ad|bc
    vals = [s1, s2, s3]
    m = min(vals)
    winners = [i for i, v in enumerate(vals) if abs(v - m) <= tol]
    if len(winners) == 1:
        return winners[0]  # 0,1,2
    return -1  # unresolved/tie

def quartet_distance(tree1, tree2, tol=1e-10):
    labels1 = {tx.label for tx in tree1.taxon_namespace}
    labels2 = {tx.label for tx in tree2.taxon_namespace}
    common = sorted(labels1 & labels2)
    n = len(common)
    if n < 4:
        raise ValueError("Need at least 4 shared taxa.")

    t1_map = {tx.label: tx for tx in tree1.taxon_namespace if tx.label in common}
    t2_map = {tx.label: tx for tx in tree2.taxon_namespace if tx.label in common}
    pdm1 = tree1.phylogenetic_distance_matrix()
    pdm2 = tree2.phylogenetic_distance_matrix()

    total = math.comb(n, 4)
    disagree = 0
    resolved_both = 0
    unresolved_any = 0

    for a, b, c, d in itertools.combinations(common, 4):
        q1 = quartet_split_id(pdm1, t1_map[a], t1_map[b], t1_map[c], t1_map[d], tol=tol)
        q2 = quartet_split_id(pdm2, t2_map[a], t2_map[b], t2_map[c], t2_map[d], tol=tol)

        if q1 == -1 or q2 == -1:
            unresolved_any += 1
            continue
        resolved_both += 1
        if q1 != q2:
            disagree += 1

    # Two normalizations:
    # 1) over all quartets
    qd_all = disagree / total
    # 2) over quartets resolved in both trees
    qd_resolved = (disagree / resolved_both) if resolved_both > 0 else float("nan")

    return {
        "shared_taxa": n,
        "total_quartets": total,
        "resolved_in_both": resolved_both,
        "unresolved_in_any": unresolved_any,
        "disagree_count": disagree,
        "quartet_distance_all": qd_all,
        "quartet_distance_resolved_only": qd_resolved,
    }
```

```python
import math
import itertools
import dendropy

def load_tree(path, schema, taxa):
    t = dendropy.Tree.get(
        path=path,
        schema=schema,
        taxon_namespace=taxa,
        preserve_underscores=True,  # important
    )
    t.is_rooted = False
    return t

def quartet_split_id(pdm, ta, tb, tc, td, tol=1e-10):
    s1 = pdm.distance(ta, tb) + pdm.distance(tc, td)  # ab|cd
    s2 = pdm.distance(ta, tc) + pdm.distance(tb, td)  # ac|bd
    s3 = pdm.distance(ta, td) + pdm.distance(tb, tc)  # ad|bc
    vals = [s1, s2, s3]
    m = min(vals)
    winners = [i for i, v in enumerate(vals) if abs(v - m) <= tol]
    return winners[0] if len(winners) == 1 else -1

def leaf_taxon_map(tree):
    # only taxa present in this tree
    return {nd.taxon.label: nd.taxon for nd in tree.leaf_node_iter() if nd.taxon is not None}

def quartet_distance(tree1, tree2, tol=1e-10):
    t1_map = leaf_taxon_map(tree1)
    t2_map = leaf_taxon_map(tree2)

    common = sorted(set(t1_map.keys()) & set(t2_map.keys()))
    n = len(common)
    if n < 4:
        raise ValueError(f"Need at least 4 shared taxa, got {n}.")

    pdm1 = tree1.phylogenetic_distance_matrix()
    pdm2 = tree2.phylogenetic_distance_matrix()

    total = math.comb(n, 4)
    disagree = 0
    resolved_both = 0
    unresolved_any = 0

    for a, b, c, d in itertools.combinations(common, 4):
        q1 = quartet_split_id(pdm1, t1_map[a], t1_map[b], t1_map[c], t1_map[d], tol=tol)
        q2 = quartet_split_id(pdm2, t2_map[a], t2_map[b], t2_map[c], t2_map[d], tol=tol)

        if q1 == -1 or q2 == -1:
            unresolved_any += 1
            continue
        resolved_both += 1
        if q1 != q2:
            disagree += 1

    return {
        "shared_taxa": n,
        "total_quartets": total,
        "resolved_in_both": resolved_both,
        "unresolved_in_any": unresolved_any,
        "disagree_count": disagree,
        "quartet_distance_all": disagree / total,
        "quartet_distance_resolved_only": (disagree / resolved_both) if resolved_both > 0 else float("nan"),
    }

# paths
sampled_nexus_path = "results_full/run_20260210_141401_ntaxa27/sampled_trees.nex"
ground_truth_newick_path = "data/ds1_best_mrbayes_cleaned_tree.nwk"

taxa = dendropy.TaxonNamespace()
trained = load_tree(trained_nexus_path, "nexus", taxa)
ground = load_tree(ground_truth_newick_path, "newick", taxa)

res = quartet_distance(trained, ground)
print(res)
```

```python
import math
from collections import Counter, defaultdict
import dendropy

def topology_key_unrooted(tree):
    tree.is_rooted = False
    tree.encode_bipartitions()
    full_mask = tree.taxon_namespace.all_taxa_bitmask()
    splits = []
    for b in tree.bipartition_encoding:
        if b.is_trivial():
            continue
        x = b.split_bitmask
        x = min(x, full_mask ^ x)
        splits.append(x)
    return tuple(sorted(splits))

def normalize(d):
    s = sum(d.values())
    return {k: v / s for k, v in d.items()} if s > 0 else {}

def tv_distance(p, q):
    keys = set(p) | set(q)
    return 0.5 * sum(abs(p.get(k, 0.0) - q.get(k, 0.0)) for k in keys)

def js_divergence(p, q):
    keys = set(p) | set(q)
    m = {k: 0.5*(p.get(k,0.0)+q.get(k,0.0)) for k in keys}
    def kl(a,b):
        out = 0.0
        for k in keys:
            av, bv = a.get(k,0.0), b.get(k,0.0)
            if av > 0 and bv > 0:
                out += av * math.log2(av/bv)
        return out
    return 0.5*kl(p,m) + 0.5*kl(q,m)

def overlap_mass(p, q):
    keys = set(p) | set(q)
    return sum(min(p.get(k,0.0), q.get(k,0.0)) for k in keys)


sampled_nexus_path = "results_full/run_20260210_141401_ntaxa27/sampled_trees.nex"
trprobs_path = "data/DS1.trprobs"

taxa = dendropy.TaxonNamespace()

# Your sampled trees
sampled = dendropy.TreeList.get(
    path=sampled_nexus_path,
    schema="nexus",
    taxon_namespace=taxa,
    preserve_underscores=True,
)
p_counts = Counter(topology_key_unrooted(t) for t in sampled)
P = normalize(p_counts)

# MrBayes trprobs (NEXUS)
mb = dendropy.TreeList.get(
    path=trprobs_path,
    schema="nexus",
    taxon_namespace=taxa,
    preserve_underscores=True,
)

# weight from tree.weight if present, else 1
q_counts = defaultdict(float)
for t in mb:
    w = float(t.weight) if getattr(t, "weight", None) is not None else 1.0
    q_counts[topology_key_unrooted(t)] += w
Q = normalize(q_counts)

print("unique sampled topologies:", len(P))
print("unique mrbayes topologies:", len(Q))
print("TV distance:", tv_distance(P, Q))
print("JS divergence (bits):", js_divergence(P, Q))
print("Overlap mass:", overlap_mass(P, Q))
```

```python
# Cell 1: load trees + helpers
import math
import random
from collections import Counter, defaultdict
import dendropy
from dendropy.calculate import treecompare

sampled_nexus_path = "results_full/run_20260211_222557_ntaxa27/sampled_trees.nex"
trprobs_path = "data/DS1.trprobs"

def canon(x):
    return x.strip().strip("'\"").replace(" ", "_")

def relabel_tree_inplace(t):
    for nd in t.leaf_node_iter():
        if nd.taxon is not None and nd.taxon.label is not None:
            nd.taxon.label = canon(nd.taxon.label)

def leaf_labels(t):
    return {nd.taxon.label for nd in t.leaf_node_iter() if nd.taxon is not None}

def normalize(d):
    s = sum(d.values())
    return {k: v/s for k, v in d.items()} if s > 0 else {}

def tv_distance(p, q):
    keys = set(p) | set(q)
    return 0.5 * sum(abs(p.get(k,0.0)-q.get(k,0.0)) for k in keys)

def js_divergence_bits(p, q):
    keys = set(p) | set(q)
    m = {k: 0.5*(p.get(k,0.0)+q.get(k,0.0)) for k in keys}
    def kl(a,b):
        out = 0.0
        for k in keys:
            av, bv = a.get(k,0.0), b.get(k,0.0)
            if av > 0 and bv > 0:
                out += av * math.log2(av/bv)
        return out
    return 0.5*kl(p,m) + 0.5*kl(q,m)

taxa = dendropy.TaxonNamespace()
sampled = dendropy.TreeList.get(path=sampled_nexus_path, schema="nexus", taxon_namespace=taxa, preserve_underscores=True)
mb = dendropy.TreeList.get(path=trprobs_path, schema="nexus", taxon_namespace=taxa, preserve_underscores=True)

for t in sampled:
    t.is_rooted = False
    relabel_tree_inplace(t)
for t in mb:
    t.is_rooted = False
    relabel_tree_inplace(t)

# shared taxa set (should be 27)
shared = leaf_labels(sampled[0]) & leaf_labels(mb[0])
print("shared taxa:", len(shared))
```

```python
from ete3 import Tree

def compare_trees_robust(tree_file1, tree_file2):
    # format=1: flexible loading that supports internal node names and branch lengths
    # format=0: default (strict)
    # format=5: supports internal node names + branch lengths + leaf names
    try:
        t1 = Tree(tree_file1, format=1)
        t2 = Tree(tree_file2, format=1)
        
        # Calculate Robinson-Foulds distance
        rf, rf_max, common_attrs, edges_t1, edges_t2, discarded_t1, discarded_t2 = t1.robinson_foulds(t2, unrooted_trees=True)
        
        print(f"RF Distance: {rf}")
        print(f"Normalized RF: {rf / rf_max if rf_max > 0 else 0}")
    except Exception as e:
        print(f"Error loading trees: {e}")

# Example usage
# compare_trees_robust("trained_tree.nwk", "true_tree.nwk")
```

```python
import dendropy

def compare_dendropy(file1, file2):
    # Load a common taxon namespace to ensure leaves are matched correctly
    tns = dendropy.TaxonNamespace()
    t1 = dendropy.Tree.get(path=file1, schema="newick", taxon_namespace=tns)
    t2 = dendropy.Tree.get(path=file2, schema="newick", taxon_namespace=tns)
    
    # Encode bipartitions to enable distance calculation
    t1.encode_bipartitions()
    t2.encode_bipartitions()
    
    # Calculate symmetric difference (Robinson-Foulds)
    distance = dendropy.calculate.tree_distance.robinson_foulds_distance(t1, t2)
    print(f"RF Distance: {distance}")
```

```python
compare_dendropy("data/nj_tree_seq.nwk", "data/ds1_best_mrbayes.nwk")
```

```python
from ete3 import Tree

# If your variables are file paths:
tree_a = Tree("data/ds1_sysnthetic_iqtree_ML.nwk", format=1)
tree_b = Tree("data/nj_tree_DS1.nwk", format=1)
tree_c = Tree("data/ds1_best_mrbayes.nwk", format=1)

# Now the comparison will work:
rf, rf_max, common_attrs, _, _, _, _ = tree_a.robinson_foulds(tree_b, unrooted_trees=True)
print(f"Comparison A-B used {len(common_attrs)} taxa. RF: {rf}")

rf2, rf_max2, common_attrs2, _, _, _, _ = tree_b.robinson_foulds(tree_c, unrooted_trees=True)
print(f"Comparison B-C used {len(common_attrs2)} taxa. RF: {rf2}")
```

```python
import dendropy
from dendropy.calculate import treecompare

# paths to your two .nwk files
tree1_path = "data/ds1_sysnthetic_iqtree_ML.nwk"
tree2_path = "data/ds1_best_mrbayes.nwk"

taxa = dendropy.TaxonNamespace()

t1 = dendropy.Tree.get(
    path=tree1_path,
    schema="newick",
    taxon_namespace=taxa,
    preserve_underscores=True,
)

t2 = dendropy.Tree.get(
    path=tree2_path,
    schema="newick",
    taxon_namespace=taxa,
    preserve_underscores=True,
)

# unrooted RF
t1.is_rooted = False
t2.is_rooted = False

# sanity check: shared leaves
lab1 = {nd.taxon.label for nd in t1.leaf_node_iter() if nd.taxon is not None}
lab2 = {nd.taxon.label for nd in t2.leaf_node_iter() if nd.taxon is not None}
shared = lab1 & lab2
print("leaves tree1:", len(lab1))
print("leaves tree2:", len(lab2))
print("shared leaves:", len(shared))
print("only tree1 (first 10):", sorted(lab1 - lab2)[:10])
print("only tree2 (first 10):", sorted(lab2 - lab1)[:10])

t1.encode_bipartitions()
t2.encode_bipartitions()

rf = treecompare.symmetric_difference(t1, t2)

# normalized RF for unrooted *binary* trees on n taxa
n = len(shared)
max_rf = 2 * (n - 3) if n >= 3 else float("nan")
norm_rf = rf / max_rf if max_rf and max_rf > 0 else float("nan")

print(f"Taxa (shared): {n}")
print(f"RF distance: {rf}")
print(f"Max RF (2*(n-3)): {max_rf}")
print(f"Normalized RF: {norm_rf:.4f}")
```
